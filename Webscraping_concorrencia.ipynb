```python
# Acessando a p√°gina:
navegador = webdriver.Chrome(service=servico) #Criando o navegador e identificando o chromedriver
navegador.get('https://www.ifood.com.br')
```

```python
# Definindo Produto:
produto = navegador.find_element('xpath','//*[@id="__next"]/div[1]/header/div[2]/form/div/input')
produto.send_keys('Hamb√∫rguer')
sleep(1)
produto.send_keys(Keys.ENTER)
```

```python
# Scraping:
elementos = navegador.find_elements('xpath','//*[@id="marmita-panel0-1"]/section/section/article/section[1]/div/div/div[*]/a/div')# Nome do arquivo txt onde voc√™ deseja salvar os resultados
nome_arquivo = 'saida_scraping.txt'

with open(nome_arquivo, 'w', encoding='utf-8') as arquivo_saida:
    for elemento in elementos:
        textos = elemento.get_attribute('innerText')
        arquivo_saida.write(textos + '\n')
    print("Dados foram gravados no arquivo com sucesso!")
```

## Primeira etapa de tratamento:

## Segunda etapa de tratamento:
        A informa√ß√£o que consta no site nem sempre vem completa,
    acontecendo de, em alguns casos, faltar o rank da lanchonete.
    Na falta de tempo para elaborar um "passo a passo" (algor√≠tmo)
    para lidar com essas informa√ß√µes faltantes, a corre√ß√£o foi feita
    de forma manual. Por isso do "checkpoint" acima, para ir corrigindo
    e checando.

```python
# Checkpoint:
arquivo = 'concorrentes.txt'
```

```python
# Lendo o arquivo com a codifica√ß√£o 'utf-8'
with open(arquivo, 'r', encoding='utf-8') as arq:
    linhas = arq.readlines()  # Usamos read() para ler todo o conte√∫do do arquivo em uma √∫nica string
    
valores = []
for linha in linhas:
    linha = linha.replace('\n','')
    valores.append(linha)
    
count = 0
for i in range(0,600,6):
    try:
        print(f'[{i+1}] {valores[i]}')
    except:
        print('fim')
```

## Constru√ß√£o do DataFrame:
        O objetivo nesse ponto √© construir o dataframe de modo que
    seja importado no Power BI sem problemas adicionais. O objetivo
    final desses dados √© serem utilizados numa ferramento da visualiza√ß√£o.

```python
# Separando as informa√ß√µes de cada estabelecimento:
matriz_dados = [(valores[i], valores[i+1], valores[i+2],
                 valores[i+5]) for i in range(0, 600, 6) if i+5 < len(valores)]

# Modelando os dados para o DataFrame:
concor_dict = {
    'Estabelecimento': [i[0] for i in matriz_dados],
    'Rank': [i[1] for i in matriz_dados],
    'Categoria': [i[2] for i in matriz_dados],
    'Frete': [i[3] for i in matriz_dados]
}

# Transformando em DataFrame:
df = pd.DataFrame(concor_dict)
df
```

## Terceira etapa de tratamento:
        Todos os valores aqui vieram de um arquivo txt, ou seja,
    s√£o todos strings. N√£o quero trabalhar com numeros sendo
    interprtados como string, deste modo, algumas convers√µes
    precisam ser feitas.

```python
# Corre√ß√£o de elementos estranhos:
df['Frete'] = df['Frete'].str.replace('R\$', '').str.replace(',', '.')
```

```python
# corre√ß√£o do tipo de dado:
for var in df.columns:
    for i in range(0,len(df)):
        try:
            float(df[var][i])
            print(f'{df[var][i]} corrigido')
        except:
            print(f'string, {df[var][i]}')
```
